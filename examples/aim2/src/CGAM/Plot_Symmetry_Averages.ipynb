```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def plot_feature_bars(data_path, feature=None, speed='FV'):
    """
    Load matchedCycles data and plot bar graphs of feature means (±SD) across interventions.
    """
    df = pd.read_csv(data_path)
    df = df[df['Speed'] == speed]

    meta = ['Subject','Intervention','Speed','Trial','Cycle','Side']
    feature_cols = [c for c in df.columns 
                    if c not in meta and pd.api.types.is_numeric_dtype(df[c])]

    if feature:
        if feature in df.columns and pd.api.types.is_numeric_dtype(df[feature]) and feature not in meta:
            feature_cols = [feature]
        else:
            print(f"Feature '{feature}' not found or not numeric.")
            return

    save_dir = r"Y:\LabMembers\Ameen\Stroke-R01-Aim-2\SavedOutcomesAim2\Overground_EMG_Kinematics\2_FromPython_CGAM\Plots"
    os.makedirs(save_dir, exist_ok=True)

    order = ['BL','MID18','MID24','POST18','POST24','MO1FU','MO3FU']

    for feat in feature_cols:
        summary = df.groupby('Intervention', as_index=False)[feat].agg(['mean','std']).reset_index()
        summary['Intervention'] = pd.Categorical(summary['Intervention'], categories=order, ordered=True)
        summary = summary.sort_values('Intervention')

        summary = summary[summary['Intervention'].notna()]  # Drop unavailable interventions
        if summary.empty:
            continue

        interventions = summary['Intervention']
        means = summary['mean']
        stds = summary['std']

        plt.figure(figsize=(8, 5))
        plt.bar(interventions, means, yerr=stds, capsize=4, color='steelblue', alpha=0.8)
        plt.title(f"{feat} ({speed})")
        plt.ylabel('Mean ± SD')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()

        save_path = os.path.join(save_dir, f"{feat}_{speed}.png")
        plt.savefig(save_path, dpi=300)
        plt.close()
        print(f"Saved: {save_path}")
```

```python
data_path = r"Y:\LabMembers\Ameen\Stroke-R01-Aim-2\SavedOutcomesAim2\Overground_EMG_Kinematics\2_FromPython_CGAM\matchedCycles.csv"

# Plot all features at FV
plot_feature_bars(data_path, speed='FV')

# # Plot one specific feature at SSV
# plot_feature_bars(data_path, feature='KNEE_RMS_JointAngles_Sym', speed='SSV')
```
